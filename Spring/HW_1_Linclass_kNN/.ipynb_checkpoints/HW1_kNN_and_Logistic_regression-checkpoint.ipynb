```python
# Ваш код здесь

def plot_hist(data_matrix, n_rows=11, n_cols=5):
    n_bins = 10

    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, 20))
    ax = axes.flatten()

    for i in range(data_matrix.shape[1]):
        ax[i].hist(data_matrix[:,i], n_bins, color="blue")
        ax[i].legend(prop={'size': 10})
        ax[i].set_title(str(i))

    fig.tight_layout()
    plt.show()

plot_hist(train_feature_matrix)
```

#### 2.2 Многоклассовая классификация: логистическая регрессия (40%)

**Комментарий:**

*Логистическая регрессия позволяет решать и задачу многоклассовой классификации. Класс ``LogisticRegression`` позвляет это делать двумя способами:* 
- *1. Стандартный One vs Rest (т.е. каждый класс отделяется от всех других). Параметр `multi_class='ovr'`.*
- *2. Используюя кросс-энтропию (оценивается сразу вектор вероятностей принадлежности классам). Параметр `multi_class='multinomial'`.*

*По умолчанию будем использовать второй способ.*


Теперь будем работать со всеми 7 типами покрытия (данные уже находятся в переменных `feature_matrix` и `labels`, если Вы их не переопределили). Разделите выборку на обучающую и тестовую с помощью метода `train_test_split`, используйте значения параметров `test_size=0.2`, `random_state=42`. Обучите логистическую регрессию  на данном датасете.

```python
import warnings
warnings.filterwarnings('ignore')

lr_clf = LogisticRegression()

train_feature_matrix, test_feature_matrix, train_labels, test_labels =\
train_test_split(feature_matrix, labels, 
                                     test_size=0.2, 
                                     random_state=42)

scaler.fit(train_feature_matrix, train_labels)
train_feature_matrix_scaled = scaler.transform(train_feature_matrix)
test_feature_matrix_scaled  = scaler.transform(test_feature_matrix)

from sklearn.metrics import f1_score, accuracy_score
lr_clf = LogisticRegression(solver='sag', max_iter=10000)
lr_clf.fit(train_feature_matrix_scaled, train_labels)

lr_clf_predicted = lr_clf.predict(test_feature_matrix_scaled)
accuracy_score = accuracy_score(lr_clf_predicted, test_labels)

print("Logistic regression: accuracy={:04f}".format(accuracy_score))
```

С помощью 5-кратной кросс-валидации (`GridSearchCV`) подберите оптимальное значение параметров `С` и `penalty`. Для параметра `C` используйте набор значений `np.linspace(0.01, 1, 10)`, а для `penalty` - `['l1', 'l2']`. Критерий качества `scoring = 'accuracy'`.

```python
# Ваш код здесь

C = np.linspace(0.01, 1, 10)

param_grid_l1 = {'C': C, 'solver':['saga']}
param_grid_l2 = {'C': C, 'solver':['lbfgs']}

cv = StratifiedKFold(train_labels, n_folds=5, shuffle=True, random_state=0)

est_l1 = LogisticRegression(penalty='l1', max_iter=10000)
est_l2 = LogisticRegression(penalty='l2', max_iter=10000)


opt_l1 = GridSearchCV(est_l1, param_grid_l1, scoring = 'accuracy', cv=cv)
opt_l2 = GridSearchCV(est_l2, param_grid_l2, scoring = 'accuracy', cv=cv)

opt_l1.fit(train_feature_matrix_scaled, train_labels)
opt_l2.fit(train_feature_matrix_scaled, train_labels)
```

Постройте графики средних значений критерия качества `accuracy` для типов регуляризации `l1` и `l2` в зависимости от коэффициента регуляризации `C`.

```python
# Ваш код здесь

def plot_scores_2 (optimizer1, optimizer2, title):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
    #optimizer2_mean
    scores1 = [[item[0]['C'], item[1], 
(np.sum((item[2]-item[1])**2)/(item[2].size-1))**0.5] for item in optimizer1.grid_scores_]
    scores1 = np.array(scores1)
    ax1.plot(scores1[:,0], scores1[:,1], 'o-')
    ax1.fill_between(scores1[:,0], scores1[:,1]-scores1[:,2], 
                                  scores1[:,1]+scores1[:,2], alpha=0.3)
    ax1.set_ylabel('accuracy')
    ax1.set_xlabel('C')
    ax1.set_title(title[0])
    y_min = min(scores1[:,1]-scores1[:,2])
    y_max = max(scores1[:,1]+scores1[:,2])
    ax1.set_ylim(y_min, y_max)
    
    scores2 = [[item[0]['C'], item[1], 
(np.sum((item[2]-item[1])**2)/(item[2].size-1))**0.5] for item in optimizer2.grid_scores_]
    scores2 = np.array(scores2)
    ax2.plot(scores2[:,0], scores2[:,1], 'o-')
    ax2.fill_between(scores2[:,0], scores2[:,1]-scores2[:,2], 
                                  scores2[:,1]+scores2[:,2], alpha=0.3)
    ax2.set_ylabel('accuracy')
    ax2.set_xlabel('C')
    ax2.set_title(title[1])
    y_min = min(scores2[:,1]-scores2[:,2])
    y_max = max(scores2[:,1]+scores2[:,2])
    ax2.set_ylim(y_min, y_max)
    plt.show()
    
plot_scores_2(opt_l1, opt_l2,["L1 regularization", "L2 regularization"])
```

```python
def plot_scores_1 (optimizer1, optimizer2, labels):
    fig  = plt.figure(figsize=(10,7))
    #optimizer2_mean
    scores1 = [[item[0]['C'], item[1], 
(np.sum((item[2]-item[1])**2)/(item[2].size-1))**0.5] for item in optimizer1.grid_scores_]
    scores1 = np.array(scores1)
    plt.plot(scores1[:,0], scores1[:,1], 'o-', color='blue', label=labels[0])
    plt.fill_between(scores1[:,0], scores1[:,1]-scores1[:,2], 
                                  scores1[:,1]+scores1[:,2], color='blue', alpha=0.3)
    scores2 = [[item[0]['C'], item[1], 
(np.sum((item[2]-item[1])**2)/(item[2].size-1))**0.5] for item in optimizer2.grid_scores_]
    scores2 = np.array(scores2)
    plt.plot(scores2[:,0], scores2[:,1], 'o-', color='red', label=labels[1])
    plt.fill_between(scores2[:,0], scores2[:,1]-scores2[:,2], 
                                  scores2[:,1]+scores2[:,2], color='red', alpha=0.3)

    plt.ylabel('accuracy')
    plt.xlabel('C')
    plt.title("L1 and L2 both")
    y_min = min (min(scores2[:,1]-scores2[:,2]), min(scores1[:,1]-scores1[:,2]))
    y_max = max(max(scores2[:,1]+scores2[:,2]), max(scores1[:,1]+scores1[:,2]))
    
    plt.ylim(y_min, y_max)
    plt.legend(loc = "lower right")
    plt.show()
plot_scores_1(opt_l1, opt_l2,["L1", "L2"])
```

#### 2.3 Многоклассовая классификация: kNN (30%).

Вспомним первое занятие. Обучите на том же датасете `KNeighborsClassifier` из `sklearn`.

```python
from sklearn.neighbors import KNeighborsClassifier
```

Попробуйте использовать различные стратегии вычисления весов: `[‘uniform’, ‘distance’]`. Выберите одно из них самостоятельно. Чем обоснован Ваш выбор?

Переберите по сетке от `1` до `10` параметр числа соседей. Покажите на графике, как менялись значения критерия качества `accuracy` при изменении числа соседей.

```python
# Ваш код здесь
from sklearn.pipeline import make_pipeline
import itertools
import tqdm

N = 11

result = np.zeros((N, 2))

kfold = sklearn.model_selection.KFold(5, shuffle=True)

for N_pair, W_pair in tqdm.tqdm(itertools.product(enumerate(range(1, N)), enumerate(['uniform', 'distance'])), total=N*2-2):
     
    result[N_pair[0]][W_pair[0]] = sklearn.model_selection.cross_val_score(
         make_pipeline(StandardScaler(), 
                                        KNeighborsClassifier(n_neighbors=N_pair[1], weights=W_pair[1])),
         train_feature_matrix,
         train_labels,
         scoring='accuracy',
         cv=kfold,
         n_jobs=-1
     ).mean()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
ax.grid()

ax.plot(range(1, N), result[:-1, 0],'o-', label='uniform weight strategy', c='blue')
ax.plot(range(1, N), result[:-1, 1],'o-', label='distance weight strategy', c='red')

ax.set_xlabel('N neighbors')
ax.set_ylabel('Mean accuracy on CV')

ax.legend()
plt.show()
```

$\textbf{Вывод:}$ distance weight strategy с числом соседей равным 4 дала наибольший accuracy, поэтому далее будем ее использовать

Используя найденное оптимальное число соседей, вычислите вероятности принадлежности к классам для тестовой выборки (`.predict_proba`).

```python
# Ваш код здесь

best_knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=4, weights='distance')
best_knn.fit(train_feature_matrix_scaled, train_labels)
predicted = best_knn.predict_proba(test_feature_matrix_scaled)

print('predicted proba mean: ', np.round (predicted.mean(axis=0), decimals=4 ))
est_prob = np.array([(train_labels == (i)).sum() for i in range(1, 8)]) / train_labels.shape[0]
print('estimated proba mean: ', np.round (est_prob, decimals=4))
```
